```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate non-stationary series with a trend
np.random.seed(0)
time = np.arange(100)
series = 0.5 * time + np.random.normal(size=100)

plt.plot(time, series, label='Non-Stationary Series')
plt.title('Non-Stationary Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

# Converting non stationarity to stationary data

```python
# First-order differencing
differenced_series = np.diff(series)

plt.plot(time[1:], differenced_series, label='First-order Differenced Series')
plt.title('Stationary Time Series (After Differencing)')
plt.xlabel('Time')
plt.ylabel('Differenced Value')
plt.legend()
plt.show()
```

```python
# Generate non-stationary series with increasing variance
series_with_variance = np.exp(0.03 * time) * (1 + np.random.normal(size=100))

# Log transformation
log_transformed_series = np.log(series_with_variance)

plt.plot(time, series_with_variance, label='Non-Stationary Series with Increasing Variance')
plt.plot(time, log_transformed_series, label='Log Transformed Series')
plt.title('Log Transformation to Reduce Variance')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
