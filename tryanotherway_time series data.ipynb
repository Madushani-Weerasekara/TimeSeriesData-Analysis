## Time Resampling

```python
df_tsla.head()
```

```python
df_tsla.resample(rule='A')
```

```python
df_tsla.resample(rule='A').min()
```

```python
# Rule A means year end frequency
df_tsla.resample(rule='A').max()['Open'].plot()
```

```python
# quartely start frequency
df_tsla.resample(rule='QS').max()['High'].plot()
```

```python
# Business end frequency
df_tsla.resample(rule='BA').max()
```

```python
# FutureWarning: 'BA' is deprecated and will be removed in a future version, please use 'BYE' instead.

df_tsla.resample(rule='BA').max().plot()
```

```python
# Business quaters
df_tsla.resample(rule='BQS').max()
```

```python
## Plotting
# FutureWarning: 'BA' is deprecated and will be removed in a future version, please use 'BYE' instead.

df_tsla['Open'].resample(rule='BYE').mean().plot(kind='bar')
```

```python
# monthly 
# FutureWarning: 'M' is deprecated and will be removed in a future version, please use 'ME' instead
df_tsla['Open'].resample(rule='ME').mean().plot(kind='bar', figsize=(16,4))
```

```python
df_tsla['Open'].resample(rule='M').mean()
```
